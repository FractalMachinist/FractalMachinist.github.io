```python
jsw = get_jobs_skills_weights(enforce_bookmarked=False)
```

```python
with open("personal_track/skipped_skills.txt", 'r') as sc_f:
    skipped_skills = set(sc_f.read().splitlines())

jsw.loc[
        jsw[("resume", "category")] == set()
    ]["skill"]\
    .query("category != 'emphasis' and category != 'company attributes' and name not in @skipped_skills")\
    .groupby(["category", "name"])\
    .agg({"share": ["sum", "mean", "count"]})\
    .sort_values(("share", "sum"), ascending=False)\
    .head(10)
```
